```python
#importing libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.transform import resize
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn import svm
from sklearn.metrics import accuracy_score, confusion_matrix
import pickle
```

# Untuk convert Matrix to Vector - flatten()

```python
# Preprocessing:
# 1) Resize 2) Flatten

target = []
images = []
flat_data = []

DATADIR = 'C:/Users/fiqga/Desktop/Skripsi/SVM_IKAN/Image Dataset'
CATEGORIES = ['segar', 'kurang segar', 'tidak segar']

for category in CATEGORIES:
    class_num = CATEGORIES.index(category) #Label Encoding
    path = os.path.join(DATADIR, category) #path to use images
    for img in os.listdir(path):
        img_array = imread(os.path.join(path, img))
        #print(img_array.shape) 
        #plt.imshow(img_array) To view an image
        img_resized = resize(img_array, (400,400,3))
        flat_data.append(img_resized.flatten())
        images.append(img_resized)
        target.append(class_num)
        #break To view an image

flat_data = np.array(flat_data)
target = np.array(target)
images = np.array(images)
```

```python
pd.DataFrame(flat_data).to_csv("flat.csv")
```

# Splitting data

```python
x_train, x_test, y_train, y_test = train_test_split(flat_data, target, test_size=0.25, random_state=42)
```

```python
param_grid = [
                {'C':[1,10,100,1000], 'kernel':['linear']},
                {'C':[1,10,100,1000], 'gamma':[0.001,0.0001], 'kernel':['rbf']},
]

svc = svm.SVC(probability=True)
clf = GridSearchCV(svc, param_grid)
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
accuracy_score(y_pred, y_test)
```

```python
confusion_matrix(y_pred, y_test)
```

```python
pickle.dump(clf,open('img_model.p', 'wb'))
```
